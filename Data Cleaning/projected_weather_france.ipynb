```python
# Import dependency
import pandas as pd
```

```python
# Read projected precipitation data in CSV to a DataFrame
file_path1 = "../Raw Data/France Precipitation Predict.csv"
precip_df = pd.read_csv(file_path1)
precip_df.head()
```

```python
# Rename unnamed column
precip_df = precip_df.rename(columns={'Unnamed: 0': 'Year'})
precip_df.head()
```

```python
# Create new dataframe with applicable provinces
new_precip_df = precip_df[['Year', 'Alsace', 'Aquitaine', 'Bourgogne', 'Champagne-Ardenne']]

# Rename columns to ID them as precipitation columns
new_precip_df = new_precip_df.rename(columns={'Alsace': 'Alsace Precip', 'Aquitaine': 'Aquitaine Precip', 'Bourgogne': 'Bourgogne Precip', 'Champagne-Ardenne': 'Champagne-Ardenne Precip'})

new_precip_df.head()
```

```python
# Read projected temperature data in CSV to a DataFrame
file_path2 = "../Raw Data/France Temp Predict.csv"
temp_df = pd.read_csv(file_path2)
temp_df.head()
```

```python
# Rename unnamed column
temp_df = temp_df.rename(columns={'Unnamed: 0': 'Year'})
temp_df.head()
```

```python
# Create new dataframe with applicable provinces
new_temp_df = temp_df[['Year', 'Alsace', 'Aquitaine', 'Bourgogne', 'Champagne-Ardenne']]

# Rename columns to ID them as temperature columns
new_temp_df = new_temp_df.rename(columns={'Alsace': 'Alsace Temp', 'Aquitaine': 'Aquitaine Temp', 'Bourgogne': 'Bourgogne Temp', 'Champagne-Ardenne': 'Champagne-Ardenne Temp'})

new_temp_df.head()
```

```python
# Merge the new_precip and new_temp DataFrames into one DataFrame
merge_df = pd.merge(new_precip_df, new_temp_df, on='Year')
merge_df.head()
```

```python
# Create province DataFrame with year, precipitation and temperature
alsace_df = merge_df[['Year', 'Alsace Precip', 'Alsace Temp']]

# Add a column to specify which province the precipitation and temperature data is for
alsace_df['Province'] = pd.Series(['Alsace' for x in range(len(alsace_df.index))])

# Rename the precipitation and temperature columns by replacing the province name with "Projected"
alsace_df = alsace_df.rename(columns={'Alsace Precip': 'Projected Precip', 'Alsace Temp': 'Projected Temp'})

# Reorder the columns
alsace_df = alsace_df[['Province', 'Year', 'Projected Precip', 'Projected Temp']]

alsace_df.head()
```

```python
# Create province DataFrame with year, precipitation and temperature
aquitaine_df = merge_df[['Year', 'Aquitaine Precip', 'Aquitaine Temp']]

# Add a column to specify which province the precipitation and temperature data is for
aquitaine_df['Province'] = pd.Series(['Aquitaine' for x in range(len(aquitaine_df.index))])

# Rename the precipitation and temperature columns by replacing the province name with "Projected"
aquitaine_df = aquitaine_df.rename(columns={'Aquitaine Precip': 'Projected Precip', 'Aquitaine Temp': 'Projected Temp'})

# Reorder the columns
aquitaine_df = aquitaine_df[['Province', 'Year', 'Projected Precip', 'Projected Temp']]

aquitaine_df.head()
```

```python
# Create province DataFrame with year, precipitation and temperature
bourgogne_df = merge_df[['Year', 'Bourgogne Precip', 'Bourgogne Temp']]

# Add a column to specify which province the precipitation and temperature data is for
bourgogne_df['Province'] = pd.Series(['Bourgogne' for x in range(len(bourgogne_df.index))])

# Rename the precipitation and temperature columns by replacing the province name with "Projected"
bourgogne_df = bourgogne_df.rename(columns={'Bourgogne Precip': 'Projected Precip', 'Bourgogne Temp': 'Projected Temp'})

# Reorder the columns
bourgogne_df = bourgogne_df[['Province', 'Year', 'Projected Precip', 'Projected Temp']]

bourgogne_df.head()
```

```python
# Create province DataFrame with year, precipitation and temperature
champagne_ardenne_df = merge_df[['Year', 'Champagne-Ardenne Precip', 'Champagne-Ardenne Temp']]

# Add a column to specify which province the precipitation and temperature data is for
champagne_ardenne_df['Province'] = pd.Series(['Champagne-Ardenne' for x in range(len(champagne_ardenne_df.index))])

# Rename the precipitation and temperature columns by replacing the province name with "Projected"
champagne_ardenne_df = champagne_ardenne_df.rename(columns={'Champagne-Ardenne Precip': 'Projected Precip', 'Champagne-Ardenne Temp': 'Projected Temp'})

# Reorder the columns
champagne_ardenne_df = champagne_ardenne_df[['Province', 'Year', 'Projected Precip', 'Projected Temp']]

champagne_ardenne_df.head()
```

```python
# Combine DataFrames into one
concat_df = pd.concat([alsace_df, aquitaine_df, bourgogne_df, champagne_ardenne_df])
concat_df
```

```python
# Save concat_df to CSV
concat_df.to_csv('../Raw Data/projected_weather_france.csv', index=False)
```
